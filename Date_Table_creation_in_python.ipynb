```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
start_date="01/01/2015"
end_date="01/01/2023"
date_table=pd.DataFrame(pd.date_range(start_date,end_date,freq='D'),columns=['Date'])
date_table['year']= date_table['Date'].dt.year
date_table['Month']= date_table['Date'].dt.month
date_table['quarter']= date_table['Date'].dt.quarter
date_table['Day of Week']= date_table['Date'].dt.dayofweek
date_table['Day Abb']= date_table['Date'].dt.strftime('%a')
date_table.set_index(['Date'],inplace=True)
date_table.to_csv('date_table')
date_table




```
